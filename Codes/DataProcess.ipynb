## 数据预处理

### Genres属性进行OneHot编码

```python
df_Mrating_movie = df_Mrating_movie.sample(frac=0.1)
df_title = df_Mrating_movie['title']
df_genres = df_Mrating_movie['genres']
df_genres.head()
```

```python
df_genres = pd.DataFrame(df_genres)
df_genres.columns = ['genres']
```

```python
df_genres
```

```python
genres = ["Action", "Adventure", "Animation", "Children", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western", "Others"]
for genre in genres:
    df_genres[genre] = 0
df_genres
```

```python
Nm_genres = df_genres.values
Nm_genres[0: 1,0]
```

```python
Nm_genres.shape
```

```python
df_genres.columns
```

```python
for i, element in enumerate(Nm_genres[0: ,0]):
    others = 1
    elements = element.split("|")
#     print(elements)
    for elementGenre in elements:
        if elementGenre in genres :
            index = genres.index(elementGenre)
            Nm_genres[i, index+1] = 1
            others = 0
    Nm_genres[i, 19] = others
Nm_genres        
```

```python
df_genres = pd.DataFrame(Nm_genres, columns = df_genres.columns)
df_genres
```

```python
df_genres = df_genres.drop(["genres"], axis = 1)
df_genres
```

```python
df_Mrating_movie.index = pd.Series(range(df_Mrating_movie.shape[0]))
```

```python
df_Mrating_movie
```

```python
df_oneHot = pd.concat([df_Mrating_movie, df_genres], axis = 1, join = "inner")
df_oneHot
```

### 备份已经OneHot编码的文件

```python
df_oneHot.to_csv('../output/B_JoinedOneHotData.csv', index = False)
```

## 映射电影上映年份

```python
import re
```

```python
Nm_title = df_title.values
Nm_title
```

```python
searchObj = re.search(r'[(][1-9][0-9][0-9][0-9][)]', Nm_title[0])
year = searchObj.group()
year = year.strip('(')
year = year.strip(')')
year
```

```python
df_tilteMap = []
for i, element in enumerate(Nm_title):
    searchObj = re.search(r'[(][1-9][0-9][0-9][0-9][)]', Nm_title[i])
    try:
        year = searchObj.group()
        year = year.strip('(')
        year = year.strip(')')
    except:
        year = 0
    df_tilteMap.append(int(year))
df_tilteMap = pd.DataFrame(df_tilteMap, columns=['titleMap'])
```

```python
df_tilteMap
```

```python
df_OH_TM = pd.concat([df_oneHot, df_tilteMap], axis = 1)
# df_OH_TM = df_OH_TM.drop('tag', axis = 1)
df_OH_TM
```

```python
df_OH_TM[genres] =  df_OH_TM[genres].astype("bool")
```

```python
df_OH_TM.info()
```

### 备份已经电影上映年份的数据

```python
df_OH_TM.to_csv('../output/C_JoinedOneHotYearData.csv', index = False)
```

## 映射评分时间

```python
Nm_ratingTime = df_OH_TM['ratingTime'].values
Nm_ratingTime
```

```python
rYear = []
rMonth = []
rDay = []
rHour = []
for i, element in enumerate(Nm_ratingTime):
    elements = element.split(" ")
    YMD = elements[0].split("-")
    Year = YMD[0]
    Month = YMD[1]
    Day = YMD[2]
    Hour = elements[1].split(":")[0]
    rYear.append(int(Year))
    rMonth.append(Month)
    rDay.append(Day)
    rHour.append(Hour)
pd_rYear = pd.DataFrame(rYear, columns=['rYear'])
pd_rMonth = pd.DataFrame(rMonth, columns=['rMonth'])
pd_rDay = pd.DataFrame(rDay, columns=['rDay'])
pd_rHour = pd.DataFrame(rHour, columns=['rHour'])
pd_rYear.info()
```

```python
pd_rMap_NonOH = pd.concat([df_OH_TM, pd_rYear.astype(int), pd_rMonth.astype(int), pd_rDay.astype(int), pd_rHour.astype(int)], axis = 1)
pd_rMap_NonOH.info()
```

### OneHot评分时间（月、日、时）

```python
pd_rMonth = pd.get_dummies(pd_rMonth).astype("bool")
pd_rMonth.head()
```

```python
pd_rDay = pd.get_dummies(pd_rDay).astype("bool")
pd_rDay.head()
```

```python
pd_rHour = pd.get_dummies(pd_rHour).astype("bool")
pd_rHour.head()
```

```python
pd_rMap = pd.concat([df_OH_TM, pd_rYear, pd_rMonth, pd_rDay, pd_rHour], axis = 1)
pd_rMap.info()
```

### 备份RatingTime的OneHot

```python
pd_rMap.to_csv('../output/D_JoinedOneHotYearRateOnehotData.csv', index = False)
pd_rMap_NonOH.to_csv('../output/E_JoinedOneHotYearRateNonOnehotData.csv', index=False)
```
